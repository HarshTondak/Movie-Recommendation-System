```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
credits.head(2)
```

```python
# Merging the movies and the credits data files on "title" attribute
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
# Now we are going to select only those attributes that will be helpful in content based recommendations
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
# Now we will be cleaning the attributes:-genres, keywords, cast and crew
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 

# literal_eval() simplifies the text(string) into its literal value(here, it's array)
```

```python
movies.isnull().sum()
```

```python
# Removing the data with null values
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# Check for duplicate values
movies.duplicated().sum()
```

```python
# Simplifying the "genres" attribute
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
# Simplifying the "keywords" attribute
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(3)
```

```python
# For "cast" attribute we will be taking only top 3 casts from the total
def convertCast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
movies.head(3)
```

```python
# Now from the "crew" we will be taking only the director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
# Now we will be removing spaces from individual strings(ex:- Zoe Saldana -> ZoeSaldana)
# This is to improve the search results
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(3)
```

```python
# Now we will convert the "overview" attribute from a string to a list of strings(each word will be a string)
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
# Now we will combine overview, genres, keywords, cast and crew attributes into a new single attribute -> "tags"
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# making a "new" data-table and dropping the above mentioned tags from it
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head(3)
```

```python
# Now we convert the "tags" list into a single string
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head(3)
```

```python
import nltk
```

```python
# We will be stemming the "tags" attribute to handle similar words
# stemming means:- ['love', 'loving', 'lover'] will all be considered as "love"
# This will improve our recommendation system
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    temp = []
    for i in text.split():
        temp.append(ps.stem(i))
    string = " ".join(temp)
    return string
```

```python
# Applying stemming on the "tags" attribute
new['tags'] = new['tags'].apply(stem)
```

```python
new.head(3)
```

```python
# Now we will perform vectorization on the "tags" attribute
# We will convert each movie's data(tags attribute value) into a vector
# The recommendation system will use these vectors to find the closest related vectors(movies)
```

```python
# We will use the "BAG OF WORDS" technique
# Here we will combine all the "tags" values of all movies into a single string
# Then we find "N" most occuring words
```

```python
# During this vectorization we will not be considering the "Stop Words"
# Stop words->  are those words that themselves don't mean anything, they are only used for support(like is, are, in, a, etc)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# "N" is taken as 5000
# "stop words" of english language will not be considered
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
# These are the 5000 words that are most occuring
cv.vocabulary_.keys()
```

```python
# Now we are having 4806 movie vectors with 5000 dimensions(each word acts as a dimension)
# So finding the relationship between two movies is done by finding the distance between them
# We are using the "COSINE Distance"(angular distance), rather than the "Euclidean Distance" as is donot perform well for higher dimensions

# The lower the angle between two vectors is, the more they are related with each other
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# We will find the relationship(similarity) of each vecotr(movie) with all the others
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
# The similarity values will lie between 0 and 1(0 being no similarity, and 1 being highly similar)
similarity
```

```python
# We are using enumerate() here to add counter to the similarity vector values of given movie_id(0 here)
list(enumerate(similarity[0]))

# This is done so that when we sort the similarity vector values for a given movie, their id can be retained
```

```python
# This is the Recommendation function
def recommend(movie):
    # First we find the index(id) of the given movie
    index = new[new['title'] == movie].index[0]
    # Then we find the similarity vector for that movie and sort it in descending order to get the highly similar movies first
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    # We will show only top 5 values(other that movie itself)
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
# "pickle" is used to export the resultant file
import pickle
# "bz2" is used for compressing the pikle output file
import bz2
```

```python
# Exporting the results
pickle.dump(new,bz2.BZ2File('movie_list.pkl','wb'))
pickle.dump(similarity,bz2.BZ2File('similarity.pkl','wb'))
```

```python
recommend('Superman')
```
